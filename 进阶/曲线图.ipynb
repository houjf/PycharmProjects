```python
%matplotlib inline
import matplotlib.pyplot as pl
import pylab
import numpy as np

x = np.linspace(-np.pi,np.pi,256,endpoint=True)

y = np.sin(x)
y1 = np.cos(x)
pl.plot(x,y,linewidth=3,color='y',label='sin',ls='--')
pl.plot(x,y1,label='cos',marker='_',markersize='100')
pylab.title('sin and cos')
pylab.xlim(-3.0,3.0)
pylab.ylim(-1.0,1.0)
pylab.xticks([-np.pi, -np.pi/2, np.pi/2, np.pi])
pylab.yticks([-1,0,1])
t=pylab.axis()
print(t)
# pylab.axis([1,3,1,4])
pylab.axhline(0.5)
pylab.axvline()
pl.grid(0.01)

pl.legend()
pl.show()
```
